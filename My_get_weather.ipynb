```python
import fpcup
fpcup.test()
```

```python
# Setup
year = 2022
brp = fpcup.io.load_brp(year)
crop_species = "barley"
province = fpcup.geo.process_input_province("Drenthe")
brp_plots = fpcup.io.query_brp(brp, province=province, crop_species=crop_species)
brp_plots #brp plots in Drenthe. totally 5048 plots
```

```python
brp_barley = brp_plots[brp_plots["crop"] == "barley (winter)"]
```

```python
#too many points, take too much time to compute
#employe k-means to reduce the number of points
```

```python
from sklearn.cluster import KMeans

X = brp_barley[["latitude", "longitude"]]
brp_barley["cluster"] = KMeans(n_clusters=50, random_state=42).fit_predict(X)

sampled = brp_barley.groupby("cluster").apply(lambda x: x.sample(1)).reset_index(drop=True)
```

```python
sampled.to_csv("./sampled_points_barley_spring.csv", index=False)
```

```python
from osgeo import gdal
import numpy as np
```

```python
#inspect the grib file
def inspect_rh_by_band(fpath, rh_band=27):
    ds = gdal.Open(fpath, gdal.GA_ReadOnly)
    if ds is None:
        raise FileNotFoundError(f"Cannot open: {fpath}")
    print("RasterCount:", ds.RasterCount)

    if rh_band > ds.RasterCount:
        raise IndexError(f"Band {rh_band} > RasterCount {ds.RasterCount}")

    b = ds.GetRasterBand(rh_band)
    md = b.GetMetadata()  
    print("Band:", rh_band)
    print("Description:", b.GetDescription())
    print("GRIB_ELEMENT:", md.get("GRIB_ELEMENT"))
    print("GRIB_COMMENT:", md.get("GRIB_COMMENT"))
    print("GRIB_UNIT:", md.get("GRIB_UNIT"))
    print("GRIB_VALID_TIME:", md.get("GRIB_VALID_TIME"))
    print("GRIB_REF_TIME:", md.get("GRIB_REF_TIME"))

    arr = b.ReadAsArray().astype("float32")
    nodata = b.GetNoDataValue()
    if nodata is not None:
        arr[arr == nodata] = np.nan

    print("RH stats [%]: min=", np.nanmin(arr), "max=", np.nanmax(arr), "mean=", np.nanmean(arr))

inspect_rh_by_band(r"E:/dataset_month1/HA43_N20_202109010000_00000_GB", rh_band=27)
```
